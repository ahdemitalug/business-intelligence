```
#QUESTION 3: Plot min and max for actual_mean_temp, actual_min_temp and actual_max_temp using bar graph. Use blue for min and red for max.
import matplotlib.pyplot as plt
n=3
min_all= [data.iloc[:,1].min(),data.iloc[:,2].min(),data.iloc[:,3].min()]
max_all= [data.iloc[:,1].max(),data.iloc[:,2].max(),data.iloc[:,3].max()]

index = np.arange(n)
bar_width = 0.1
opacity = 0.5

rec1 = plt.bar(index,min_all, bar_width,
alpha=opacity,
color='y',
label='Minimum')

rec2 = plt.bar(index + bar_width, max_all, bar_width,
alpha=opacity,
color='g',
label='Maximum')

plt.xlabel('Temperature')
plt.ylabel('Value')
plt.title('Min_Max Scale')
plt.xticks(index + bar_width, ('mean temp', 'min temp','max temp'))
plt.legend()

plt.show()
```

```
#QUESTION 8 : Make the correlation matrix plot and scatterplot matrix taking all temperature attributes in the dataset
import pandas as pd
from pandas.plotting import scatter_matrix
names = ['actual_mean_temp', 'actual_min_temp', 'actual_max_temp', 'average_min_temp', 'average_max_temp', 'record_min_temp', 'record_max_temp']
data = data[names]
correlations = data.corr()

# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,7,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
plt.xticks(rotation=90)
ax.set_yticklabels(names)
ax.set_xticklabels(names)
plt.show()

scatter_matrix(data)
plt.show()
```
